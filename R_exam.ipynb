````r
qvec1 <- c("2020 Q4", "2021 Q1", "2021 Q2")
qvec2 <- c("Q4/20", "Q1/21", "Q2/21")
qvec3 <- c("Q4-2020", "Q1-2021", "Q2-2021")
````

Для преобразования qvec1 в объект Date мы можем использовать функцию
as.yearqtr из пакета zoo двумя способами:

````r
install.packages("zoo")
library(zoo)
as.Date(as.yearqtr(qvec1, format = "%Y Q%q"))
as.Date(as.yearqtr(qvec1, format = "%Y Q%q"), frac = 1)
````

1. Опишите результаты. В чем разница? Какой из них, по вашему мнению,
использовать предпочтительнее?


По умолчанию параметр `frac` равен 0. В таком случае при переводе квартала в дату выбирается первый соответствующий ему день. При изменении значения параметра на 1 выбираетмся последняя дата в квартале. Для различных целей могут быть полезны оба варианта. Например, если нам нужно найти i-ый день квартала, удобнее будет пользоваться первым вариантом (`frac = 0`).
