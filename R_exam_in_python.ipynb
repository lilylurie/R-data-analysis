## **Упражнение 41**. Рассмотрим точечную диаграмму первых двух главных компонент для данных по семенам (seeds):

````python
df_seeds = pd.read_csv("seeds.csv")
df_seeds.head()
````

````python
import sklearn
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = df_seeds.drop(["Variety"], axis = 1)
y = df_seeds.Variety
scaler = StandardScaler()
scaler.fit(X)
X=scaler.transform(X)

pca = PCA()
x_new = pca.fit_transform(X)

def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley, c = y)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel("PC{}".format(1))
plt.ylabel("PC{}".format(2))
plt.grid()

myplot(x_new[:,0:2],np.transpose(pca.components_[0:2, :]))
plt.show()
````

````python
props = [0.1010, 0.2546, 0.6009, 0.0400, 0.0035]

def percentage(part, whole):
  Percentage = round(100 * float(part)/float(whole), 1)
  return str(Percentage) + "%"

percents = []
for i in props:
  percents.append(percentage(i, sum(props)))
percents
````
